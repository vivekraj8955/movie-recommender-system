```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):        
        L.append(i['name'])
    return L
```

```python
convert(movies.iloc[0].genres)
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert(obj):
    L=[]
    counter=0;
    for i in ast.literal_eval(obj): 
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'].apply(convert)
```

```python
movies['cast']=movies['cast'].apply(convert)
```

```python
movies.head()
```

```python
def finddirector(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L       
```

```python
movies['crew']=movies['crew'].apply(finddirector)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew'];
```

```python
movies.head()
```

```python
newdf=movies[['movie_id','title','tags']]
newdf.head()
```

```python
newdf['tags']=newdf['tags'].apply(lambda x:" ".join(x))
```

```python
newdf.head()
```

```python
newdf['tags']=newdf['tags'].apply(lambda x:x.lower())
```

```python
newdf.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(newdf['tags']).toarray()
vectors.shape
```

```python
newdf.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
newdf['tags']=newdf['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend(movie):
    movie_index=newdf[newdf['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(newdf.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
